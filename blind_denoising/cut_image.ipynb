```python
import os
import numpy as np
import cv2
import h5py
from tqdm.notebook import tqdm
import scipy.io as scio
```

```python
cut_size = 256
folder_ori = "/root/autodl-tmp/SIDD_Small_Raw_Only/"
image_folder_ori = folder_ori + "Data/"
image_list_ori = os.listdir(image_folder_ori)
```

```python
def get_patchs(w, h, cut_size):
    size = cut_size
    patchs = []
    w_n = w//size
    h_n = h//size
    for wi in range(w_n):
        for hi in range(h_n):
            patchs.append([wi*size,(wi+1)*size,hi*size,(hi+1)*size])    # xmin,xmax,ymin,ymax
        patchs.append([wi*size,(wi+1)*size,h-1-size,h-1])
     
    for hi in range(h_n):
        patchs.append([w-1-size,w-1,hi*size,(hi+1)*size])    # xmin,xmax,ymin,ymax
    patchs.append([w-1-size,w-1,h-1-size,h-1])
    return patchs


def cut_image(image_folder, image_id, cut_size):
    #get image
    inoisy = h5py.File(image_folder + image_id + "/NOISY_RAW_010.MAT")
    igt = h5py.File(image_folder + image_id + "/GT_RAW_010.MAT")
    inoisy = np.float32(np.array(inoisy['x']).T)
    igt = np.float32(np.array(igt['x']).T)
    w_origin, h_origin = igt.shape
    
    #get patchs
    patchs = get_patchs(w_origin, h_origin, cut_size)# xmin,xmax,ymin,ymax
    
    #cut image
#     inoisy_cuts = []
#     igt_cuts = []
    for i, patch in enumerate(tqdm(patchs)):
        inoisy_cut = inoisy[patch[0]:patch[1],patch[2]:patch[3]]
        igt_cut = igt[patch[0]:patch[1],patch[2]:patch[3]]
        
        save_path_inoisy = "/root/autodl-tmp/SIDD_Small_Raw_Only/new_Data/" + image_id + '_' + str(i) + "/NOISY_RAW_010.MAT"
        save_path_igt = "/root/autodl-tmp/SIDD_Small_Raw_Only/new_Data/" + image_id + '_' + str(i) + "/GT_RAW_010.MAT"
        scio.savemat(save_path_inoisy, {'var_name':inoisy_cut})
        scio.savemat(save_path_igt, {'var_name':igt_cut})
```

```python
n = 0
for image_id in image_list_ori:
    cut_image(image_folder_ori, image_id, cut_size)
    n += 1
    if n>0: break
```
